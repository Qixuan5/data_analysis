```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 绘制柱状图
plt.figure(figsize=(18, 6))

file_path = 'beijing_result.xlsx'
df = pd.read_excel(file_path)

valid_congestion_ratio = df['拥堵比例'][7859:-2]

hours_ticks = np.arange(0, 290, 1) 

plt.bar(hours_ticks, valid_congestion_ratio, color='#FDD5A1')
plt.title('北京拥堵情况')
plt.xlabel('小时')
plt.ylabel('拥堵比例 (%)')
plt.xticks(ticks=hours_ticks, labels=[])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
plt.savefig('beijing.png', dpi=300)
```
