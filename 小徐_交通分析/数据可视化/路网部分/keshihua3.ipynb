```python
import matplotlib.pyplot as plt
import pandas as pd

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

file_path = '数据汇总.xlsx'
data = pd.read_excel(file_path, sheet_name='实时')


hours = data['时间']
shenzhen = data['深圳']
beijing = data['北京']
guangzhou = data['广州']
shanghai = data['上海']

# 创建图表
plt.figure(figsize=(10, 6))

# 绘制折线图
plt.plot(hours, shenzhen, label='深圳', color='#84D9F9')
plt.plot(hours, beijing, label='北京', color='#FDD5A1')
plt.plot(hours, guangzhou, label='广州', color='#8EF174')
plt.plot(hours, shanghai, label='上海', color='#FDCAFA')

# 添加标题和标签
plt.title('单日实时拥堵情况')
plt.xlabel('小时')
plt.ylabel('拥堵百分比')

# 添加图例
plt.legend()

# 显示图表
plt.show()
```
