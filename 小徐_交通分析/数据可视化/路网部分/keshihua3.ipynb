```python
import matplotlib.pyplot as plt
import pandas as pd

# 
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False  # 

file_path = '.xlsx'
data = pd.read_excel(file_path, sheet_name='')


hours = data['']
shenzhen = data['']
beijing = data['']
guangzhou = data['']
shanghai = data['']

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(hours, shenzhen, label='', color='#84D9F9')
plt.plot(hours, beijing, label='', color='#FDD5A1')
plt.plot(hours, guangzhou, label='', color='#8EF174')
plt.plot(hours, shanghai, label='', color='#FDCAFA')

# 
plt.title('')
plt.xlabel('')
plt.ylabel('')

# 
plt.legend()

# 
plt.show()
```
