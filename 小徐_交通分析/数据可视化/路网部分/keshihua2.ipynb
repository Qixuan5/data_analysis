```python
import matplotlib.pyplot as plt
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

cities = ['北京', '广州', '上海', '深圳']
traffic_data = {
    '畅行': [beijing_1, guangzhou_1, shanghai_1, shenzhen_1],
    '缓行': [beijing_2, guangzhou_2, shanghai_2, shenzhen_2],
    '拥堵': [beijing_3, guangzhou_3, shanghai_3, shenzhen_3],
    '严重拥堵': [beijing_4, guangzhou_4, shanghai_4, shenzhen_4]
}

# 准备数据
categories = list(traffic_data.keys())
values = np.array(list(traffic_data.values()))

# 设置颜色
colors = ['#4CAF50', '#FFC107', '#F44336', '#8B0000']

# 创建图表
fig, ax = plt.subplots(figsize=(6, 3))

# 绘制堆积柱状图
bottom = np.zeros(len(cities))  # 初始底部高度为0
for i, category in enumerate(categories):
    bars = ax.bar(cities, values[i], bottom=bottom, label=category,\
                                                  color=colors[i])
    bottom += values[i]  # 更新底部高度

# 添加标题和标签
ax.set_xlabel('城市')
ax.set_ylabel('百分比')
ax.set_title('不同城市的交通状况对比')
fig.tight_layout()

# 显示图表
plt.show()
```
