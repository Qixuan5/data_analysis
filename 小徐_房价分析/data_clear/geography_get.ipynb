```python
from geopy.geocoders import Nominatim
import folium
import pandas as pd
import requests
import time

# 读取原始数据
df = pd.read_csv("深圳cleaned_data.csv")

# 高德API Key
API_KEY = '841134efb8755f7025841b7ef85460c6'
# OUTPUT_FILE = 'beijing_house_map.html'

# 构建完整地址
df['地址'] = '深圳市' + df['区'] + df['街道'] + df['小区名称']

# 地理编码函数（调用高德API）
def get_location(address, key):
    url = f'https://restapi.amap.com/v3/geocode/geo?address={address}&output=json&key={key}'
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            if data['status'] == '1' and int(data['count']) > 0:
                location = data['geocodes'][0]['location']
                lon, lat = map(float, location.split(','))
                return lat, lon
    except Exception as e:
        print(f"错误: {e}")
    return None, None

# 逐行处理地址并实时更新 DataFrame 中的经纬度
print("开始使用高德地图进行地址解析...")

for idx, address in enumerate(df['地址']):
    lat, lon = get_location(address, API_KEY)
    
    # 实时更新 DataFrame 中的经纬度列
    df.at[idx, 'lat'] = lat if lat else None
    df.at[idx, 'lon'] = lon if lon else None
    
    if lat and lon:
        print(f"成功：{address} -> ({lat}, {lon})")
    else:
        print(f"失败：{address}")
    
    time.sleep(0.5)  # 控制请求频率

# 将结果保存到新的CSV文件
df.to_csv("深圳_geo.csv", index=False)
print("\n✅ 新的CSV文件（深圳_geo.csv）保存成功！")
```
